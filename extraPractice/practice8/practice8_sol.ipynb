```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup

s = Service("./chromedriver.exe")
driver = webdriver.Chrome(service=s)

#s = Service("./geckodriver.exe")
#driver = webdriver.Firefox(service=s)

driver.get("https://www.104.com.tw/jobs/main/")
time.sleep(1)

search = driver.find_element(By.ID, 'ikeyword')
search.send_keys("工程師")
time.sleep(1)

btn = driver.find_element(By.CLASS_NAME, 'btn.btn-primary.js-formCheck')
btn.click()
time.sleep(2)


page_html  = driver.page_source
print(page_html)


driver.close() 
```

```python
# use beautifulsoup to analyze html
soup = BeautifulSoup(str(page_html), 'html.parser')

job_blocks = soup.find_all("div",{'class':'b-block__left'})

with open('result.csv', 'w') as f:
    f.write('職缺名稱,公司名,薪資\n')    
    for i in job_blocks:
        job_info = i.find_all("a")
        salary = i.find("span",{'class':'b-tag--default'})
    
        if len(job_info) == 2:
            job_name = job_info[0].text.strip()
            company_name = job_info[1].text.strip()
            salary_string = salary.text.replace(',','')
            #print(job_info[0].text.strip())
            #print(job_info[1].text.strip())
            #print(salary_string)
            f.write(job_name+','+company_name+','+salary_string+'\n')
            
            f.write('')

```
